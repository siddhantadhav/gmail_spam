```python
import numpy as np
import pandas as pd
```

```python
dt = pd.read_csv('spam1.csv')
dt.head(10)
```

```python
dt['spam'] = dt['type'].map({'spam': 1,'ham': 0}).astype(int)
dt.head(5)
```

```python
def token(text):
    return text.split()
```

```python
dt['text'] = dt['text'].apply(token)
```

```python
from nltk.stem.snowball import SnowballStemmer
porter = SnowballStemmer('english',ignore_stopwords=False)
```

```python
def stemIt(text):
    return [porter.stem(word) for word in text]
```

```python
dt['text'] = dt['text'].apply(stemIt)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmit(text):
    return [lemmatizer.lemmatize(word,pos='a') for word in text]
```

```python
dt['text'] = dt['text'].apply(lemmit)
```

```python
from nltk.corpus import stopwords
stopWords = stopwords.words('english')
```

```python
def stopIt(text):
    review = [word for word in text if not word in stopWords]
    return review
```

```python
dt['text']=dt['text'].apply(stopIt)
```

```python
dt['text'] = dt['text'].apply(' '.join)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
y = dt.spam.values
x = tfid.fit_transform(dt['text'])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtext, ytrain, ytext = train_test_split(x,y,random_state = 1, test_size = 0.2, shuffle = False)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(xtrain, ytrain)
y_pred = clf.predict(xtext)
from sklearn.metrics import accuracy_score
acc_log = accuracy_score(y_pred, ytext)*100
print(f'Accuracy Log : {acc_log}')
```

```python
from sklearn.svm import LinearSVC
linearSVC = LinearSVC(random_state = 0)
linearSVC.fit(xtrain, ytrain)
y_pred = linearSVC.predict(xtext)
acc_linear = accuracy_score(y_pred, ytext)*100
print(f'Accuracy : {acc_linear}')
```
